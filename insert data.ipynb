```python
from docx import Document

document = Document('/content/drive/MyDrive/Online_FAQ_LMS_2022.docx')
help(document)
```

```python
questions=[]
answers=[]
count=0
for para in document.paragraphs:
    ans=[]
    text=str(para.text)
    if text.find('?') >0:
      questions.append(text)
      print("Questions::>>",text)
    else:
      ans.append(text)
      answers.append(ans)
      print("Answers::>>",ans)
print(count)
```
