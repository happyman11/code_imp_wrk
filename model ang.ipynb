```python
import tensorflow as tf
import numpy as np
import keras
from keras.datasets import mnist
from keras.models import Model
from keras.layers import Dense, Input
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten
from keras import backend as k
from tensorflow.keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
img_rows, img_cols=28, 28
 
if k.image_data_format() == 'channels_first':
   x_train = x_train.reshape(x_train.shape[0], 1, img_rows, img_cols)
   x_test = x_test.reshape(x_test.shape[0], 1, img_rows, img_cols)
   inpx = (1, img_rows, img_cols)
 
else:
   x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
   x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
   inpx = (img_rows, img_cols, 1)
 
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255

inpx = Input(shape=inpx)
layer1 = Conv2D(128, kernel_size=(3, 3), activation='relu')(inpx)
layer2 = Conv2D(128, (3, 3), activation='relu')(layer1)
layer3 = MaxPooling2D(pool_size=(3, 3))(layer2)
layer4 = Dropout(0.2)(layer3)
layer5 = Flatten()(layer4)
layer6 = Dense(250, activation='relu')(layer5)
layer7 = Dense(10, activation='softmax')(layer6)



model = Model([inpx], layer7)
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=keras.losses.categorical_crossentropy,
              metrics=['accuracy'])
 
model.fit(x_train, y_train, epochs=20, batch_size=64,validation_data=[x_test ,y_test])
```

```python

tf.keras.models.save_model(
    model,
    filepath="/content/model",
    overwrite=True,
    include_optimizer=True,
    save_format=None,
    signatures=None,
    options=None,
    save_traces=True
)
```

```python
def model_weight_ensemble(members, weights):

  n_layers = len(members[0].get_weights())
  print("n_number of layers::::",n_layers)
  avg_model_weights = list()
	  
  for layer in range(n_layers):
    model1=members[0].get_weights()[layer]
    print("Model 1 ::: layer number ::",layer,"shape::",model1.shape)
    model2=members[1].get_weights()[layer]
    print("Model 2 ::: layer number ::",layer,"shape::",model2.shape)
    model3=members[2].get_weights()[layer]
    print("Model 3 ::: layer number ::",layer,"shape::",model3.shape)
    # new_model1=weights[0]*model1
    # new_model2=weights[1]*model2
    # new_model3=weights[2]*model3
    # avg_layer_weights = np.sum([ new_model1, new_model2, new_model3], axis=0)
    avg_layer_weights = np.sum([model1,model2,model3], axis=0)
    avg_model_weights.append(avg_layer_weights)

  model = tf.keras.models.clone_model(members[0])
  model.set_weights(avg_model_weights)
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
 
  return model
```
