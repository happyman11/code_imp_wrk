```python
def padding(text,size):
  text_size=len(text)
  padded_text=[]
  if text_size < size:
    append_size=size-text_size
    for i in text:
      padded_text.append(i)

    for i in range(0,append_size):
      padded_text.append(0)
  return(padded_text)   

#courcera

courcera_feature_title,courcera_feature_org=[],[]

for Title,organisation in   zip(courcera_title_vectorise,courcera_org_embed):  
    courcera_feature_title.append(padding(Title,10))
    courcera_feature_org.append(padding(organisation,5))

#show
print(courcera_feature_title)


```
